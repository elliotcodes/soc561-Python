## Approach 1: Using Nested Loops
#### **&rarr;** **Open baby name files and extract `year`**

```python
# Opening files to locate `year` and extract it
for file in os.listdir():
    file= open(file, 'r')
    for line in file:
        line= line.rstrip()
        year= re.findall('Popularity in ([0-9][0-9][0-9][0-9])',line)
        if len(year) > 0:
            print(year)
```

#### &rarr; **Extract `year` and save as variable, then create dictionary and print to text file**

```python
# Opening files, extract year (slightly differently), and save as variable
with open('babyNames.txt', 'w') as babyNames_File:
    for rawfile in os.listdir():
        if not rawfile.startswith("."):
            file= open(rawfile, 'r')
            year= ""
            for line in file:
                ## remove any training characters (e.g., spaces)
                line= line.rstrip()
                searchYear= re.findall(r'Popularity in ([0-9][0-9][0-9][0-9])',line)
                if len(searchYear) > 0: ## length of 'year' > 0
                    year=searchYear[0]     
            print(year)
            
            # Create dictionary where rank is the key and names are values associated with the rank key
            dictionary= {}
            file= open(rawfile, 'r')
            for line in file:
                line = line.rstrip()
                ## string covers (rank)(name)(name) via (\d+)(\w+)(\w+)
                string= re.findall(r'<td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>', line)
                if len(string) > 0:
                    rank=string[0][0]
                    male_name=string[0][1]
                    female_name=string[0][2]
                    ## Make dict where rank=key and names are values associated w key
                    dictionary[rank]=(male_name, female_name)  
            print(dictionary)
            
            # Print year + each line of dictionary in comma-delimited file
            for key,value in dictionary.items():
                rank=key
                male_name=value[0]
                female_name=value[1]
                text= year + "," + rank + "," + male_name + "," + female_name + "\n"
                
                # should result in 4 variables with 10,000 obvs each
                babyNames_File.write(text) 
```
